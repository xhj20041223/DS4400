```python
import numpy as np

i = 0

while i < 3:
    M = np.random.randint(-10, 11, size=(3, 3))
    i += 1
    print(M)
    if np.linalg.det(M) != 0:
        try:
            M_inv = np.linalg.inv(M)
            print(M_inv)
            identity_check = np.dot(M, M_inv)
            print(np.allclose(np.dot(M, M_inv), np.eye(3)))
        except np.linalg.LinAlgError:
            print("cannot inverse")
```
